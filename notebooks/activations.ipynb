# Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from evolving_networks.activations import *
```

```python
z_container = np.arange(-10.0, 10.0, 0.2)

def plot_activation(activation, data, title):
    plt.plot(list(map(lambda z: activation(z), data)))
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

# Identity activation

```python
plot_activation(identity_activation, z_container, 'Identity activation')
```

# Binary step activation

```python
plot_activation(binary_step_activation, z_container, 'Binary step activation')
```
